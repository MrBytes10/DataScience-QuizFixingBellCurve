````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy.optimize import curve_fit

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
````

````python
steps= (200-90)/1100
print(steps)
````

````python
array1= np.arange(90,200, steps)
print(array1)
````

````python
# checking how it looks via plotting
plt.figure(figsize= (16,9), dpi=100)
plt.title("Line Plot of array1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.plot(array1)
plt.show()
````

````python
# straight lines aren't good for this, so lets create random varriance of the data
data=norm.rvs(array1)
````

````python
axis=np.linspace(1,1100,1100)
````

````python
# plot a scatter plot
plt.figure(figsize= (16,9), dpi=100)
plt.title("A Scatter Plot of the array1")
sns.set_style("darkgrid")
sns.scatterplot(axis,data)
````
